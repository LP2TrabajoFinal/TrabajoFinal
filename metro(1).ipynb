```python
#importamos lo paquetes que vamos a usar 
import requests
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
r = requests.get('https://www.metro.pe/abarrotes/arroz/arroz-extra-y-superior')
```

```python
soup = BeautifulSoup(r.content,'html.parser')
```

```python
titulos = soup.find_all('a',attrs={"class":'product-item__name'})
```

```python
titulos = [i.text for i in titulos]
```

```python
urls = soup.find_all('a',attrs={"class":'product-item__name'})
```

```python
urls = [i.get('href') for i in urls]
```

```python
tags = soup.find_all('span',{'class':'product-prices__value product-prices__value--best-price'})
precios=list()
for tag in tags:
    precios.append(tag.contents[0])
```

```python
df = pd.DataFrame({"titulo":titulos,"url":urls,"precios":precios})
```

```python
df
```
