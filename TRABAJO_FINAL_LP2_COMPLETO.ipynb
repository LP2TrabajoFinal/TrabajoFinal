```python
#importamos los paquetes que vamos a necesitar
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

from selenium import webdriver
import os 
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText
from email.encoders import encode_base64
from smtplib import SMTP
import os
```

```python
#--------------------PLAZA VEA-----------------#
ruta = os.getcwd()
print(ruta)
driver_path='{}\chromedriver.exe'.format(ruta) #DEBEMOS DESCARGAR EL CHROMEDRIVER.EXE PARA PODER USAR EL WEBDRIVER
print(driver_path)
driver= webdriver.Chrome(driver_path)
#Buscamos el url de la pagina de plaza vea para hacer el Web scraping 
driver.get('https://www.plazavea.com.pe/') #PARA QUE AUTOMATICAMENTE SE ABRA EN EL NAVEGADOR LA PAGINA PLAZA VEA
# funcion de hacer click, que este caso sera automatico 
search_bar=driver.find_element("id","search_box") 
search_bar.clear()
search_bar.send_keys("leche gloria") #ENVIARLE LA SOLICITUD DEL PRODUCTO REQUERIDO EN EL BUSCARDOR DE LA WEB DE METRO
search_bar.send_keys(Keys.RETURN)
```

```python
#PARA PLAZA VEA TRABAJAMOS CON BeautifulSoup
soup=BeautifulSoup(driver.page_source)

#LISTA CON LOS ELEMENTOS QUE SIGAN SOUP 
nombre=soup.find_all('a',{"class":"Showcase__name"})
titulo_productos=[title.text for title in nombre]

precio=soup.find_all('div',{"class":"Showcase__salePrice"})
precio_productos=[title.text for title in precio]

links=soup.find_all('a',{"class":"Showcase__name"})
links_productos=[i.get('href') for i in links]

#guardamos los datos scrapeados para guardalos en un data frame para luego hacer la comparación de los productos con 
#los otros supermercados requeridos
data={"nombre_productos":titulo_productos,
               "precio_productos":precio_productos,
               "links_productos":links_productos}
df1=pd.DataFrame(data)
#df.to_excel("PRODUCTOS_PLAZA_VEA.xlsx")
df1
```

```python
#METRO

#Buscamos el url de la pagina de metro para hacer el Web scraping al cual le asignaremos un mobre de url
url = 'https://www.metro.pe/' 
PATH='C:\Program Files (x86)\chromedriver.exe'  #DEBEMOS DESCARGAR EL CHROMEDRIVER.EXE PARA PODER USAR EL WEBDRIVER
options = webdriver.ChromeOptions()             
options.add_argument('--incognito')             #PARA QUE AL MOMENTO DE ABRIR LA NUEVA PESTAÑA DE BUSQUEDA SEA EN MODO INCOGNITO
driver=webdriver.Chrome(PATH,options=options)  
driver = webdriver.Chrome(PATH,options=options)
driver.get(url)                                 #PARA QUE AUTOMATICAMENTE SE ABRA EN EL NAVEGADOR LA PAGINA METRO EN MODO INCOGNITO

# funcion de hacer click, que este caso sera automatico 
search_bar = driver.find_element("id","search-autocomplete-input") #PARA BUSCAR EL ID DEL BUSCADOR DE LA PAGINA DE METRO PARA ENVIARLE LA SOLCITUD DEL PRODUCTO 
search_bar.clear()
search_bar.send_keys("leche gloria")          #ENVIARLE LA SOLICITUD DEL PRODUCTO REQUERIDO EN EL BUSCARDOR DE LA WEB DE METRO 
search_bar.send_keys(Keys.RETURN)            #PARA OBTENER EL RESULTADO DE LA PAGINA EN EL DRIVER 

# LISTA CON LOS ELEMENTOS QUE SIGAN EL XPATH 
titulo_productos=driver.find_elements("xpath","//a[@class='product-item__name']")
# UN ARROBA Y LUEGO SE DEFINE LA CLASE COMO SI FUERA UNA VARIABLE STRING
# SE REALIZA UN MINI-BUCLE PARA CONVERTIR A TEXTO CADA ELEMENTO
titulo_productos=[title.text for title in titulo_productos]

# LISTA CON LOS ELEMENTOS QUE SIGAN EL XPATH
precio_productos=driver.find_elements("xpath","//span[@class='product-prices__value product-prices__value--best-price']")
precio_productos=[title.text for title in precio_productos]

# LO MISMO PERO CON LOS LINKS DE CADA PRODUCTO
links_productos=driver.find_elements("xpath","//div[@class='product-item__info']//a[1]")
links_productos=[ link.get_attribute("href") for link in links_productos]

#guardamos los datos scrapeados para guardalos en un data frame para luego hacer la comparación de los productos con 
#los otros supermercados requeridos
data_productos={"nombre_productos_metro":titulo_productos,
               "precio_productos_metro":precio_productos,
               "links_productos_metro":links_productos}
df=pd.DataFrame(data_productos)
df
```

```python
#AUTOMATIZACIÓN PARA EL ENVIO DE LA COMPARACIÓN DE LOS PRODUCTOS A UN CORREO ELECTRONICO A ELECCIÓN DE FORMA AUTOMATICA CON 
#PYTHON 

emisor =input('Ingrese su coreo electronico: ') #correo electronico con el que aremos el envio
contra=input('Ingrese contraseña de su correo electronico: ') #contraseña de apliacaciones del gmail de 16 digitos 
asunto=input('ingrese el asunto del correo electronico: ') #asunto del correo a enviar
cuerpo=input('Ingrese mensaje o cuerpo del correo electronico: ') # mensaje del correo electronico
archivo=input('Ingrese el nombre del archivo a enviar con su respectiva extension: ') #con el que almacenamos los data frame con su respectiva extencion 
receptor=input('Ingrese el correo electronico receptor: ') #correo electronico del receptor

#hacer el envio al correo 
mensaje = MIMEMultipart('plain')
mensaje['From']=emisor # de quien se envia 
mensaje['To']=receptor # quien va a recibir
mensaje['subjet']=asunto
mensaje['subjet']=cuerpo
tema=MIMEText(cuerpo,'html')
mensaje.attach(tema)
adjunto=MIMEBase('application','octect-stream')
#para poder hacer el envio del archivo adjunto
adjunto.set_payload(open(archivo,'rb').read())
encode_base64(adjunto) #codificarlo para hacer un correcto envio

adjunto.add_header('content-Disposition','attachment;filename="%s"'% os.path.basename(archivo))
mensaje.attach(adjunto)
smtp=SMTP(host='smtp.gmail.com',port=587)
smtp.starttls()
smtp.login(emisor,contra) # permite ingresar nuesto email para hacer el envio respectivo 
smtp.sendmail(emisor,receptor,mensaje.as_string())
smtp.quit()
print('Correo enviado exitosamente a :',receptor)

print('= programa ejecutado y finalizado correctamente=')
```
